```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib qt


from mpl_toolkits.mplot3d import Axes3D
```

```python
#GIVEN DATA : VALUES OF X1,X1..... AND Y
x1_val=np.array([[1,2,3,4],
                 [1,2,3,5],
                 [4,5,6,7]])
print(x1_val)
```

```python
#GIVEN DATA : VALUES OF X1,X1..... AND Y

x1_val=np.array([60,70,75,55,80,65,68,62,72,67])

x2_val=np.array([10,8,5,12,6,9,7,11,4,8])

y_values=np.array([25,15,20,28,10,22,18,26,17,23])
print(len(y_values))


#constructing the LAMBDA matrix
l_x1=len(x1_val)
L = np.column_stack((np.ones(l_x1), x1_val, x2_val))
#print(L)


#NOW WE WILL USE THE FORMULA TO FIND THE COEFFICIENTS


L_Trans=np.transpose(L)

#print(L_Trans)
A=np.linalg.inv(L_Trans @L)@ L_Trans @ y_values

print(A)
l_0=A[0]
print(l_0)
l_1=A[1]
l_2=A[2]
fig=plt.figure(figsize=(8,8))

ax=fig.add_subplot(121,projection='3d')
ax.scatter(x1_val, x2_val, y_values, c='blue', marker='*', label='Data points')

x1_range=np.linspace(min(x1_val),max(x1_val),10)#ssn

x2_range = np.linspace(min(x2_val), max(x2_val), 10)

x1_mesh,x2_mesh=np.meshgrid(x1_range,x2_range)

y_mesh=l_0+l_1*x1_mesh+l_2*x2_mesh
print(y_mesh)


ax.plot_surface(x1_mesh,x2_mesh,y_mesh,alpha=0.42,cmap='autumn',label='regression plane')
plt.legend()
ax.set_xlabel('X1->X')
ax.set_ylabel('X2->Y')
ax.set_zlabel('Y->Z')
ax.set_title('MULTIPLE LINEAR REGRESSION')


```
